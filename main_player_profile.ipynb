```python


import pandas as pd
import json

# Load JSON file into a variable
with open("player_profile.json", "r", encoding="utf-8") as f:
    player_profile = json.load(f)
    player = player_profile

rows = []

for season in player.get("seasons", []):
    for team in season.get("teams", []):
        
        stats = team.get("statistics", {})
        
        row = {
            # -------- PLAYER BASIC INFO (top-level) --------
            "player_id": player.get("id"),
            "player_name": player.get("name"),
            "first_name": player.get("first_name"),
            "last_name": player.get("last_name"),
            "abbr_name": player.get("abbr_name"),
            "position": player.get("position"),
            "height": player.get("height"),
            "weight": player.get("weight"),
            "birth_place": player.get("birth_place"),
            "eligibility": player.get("eligibility"),
            "status": player.get("status"),

            # -------- SEASON INFO --------
            "season_id": season.get("id"),
            "season_year": season.get("year"),
            "season_type": season.get("type"),

            # -------- TEAM INFO --------
            "team_id": team.get("id"),
            "team_name": team.get("name"),
            "team_market": team.get("market"),
            "team_alias": team.get("alias"),

            # -------- BASIC GAME STATS --------
            "games_played": stats.get("games_played"),
            "games_started": stats.get("games_started"),

            # -------- RUSHING --------
            "rush_attempts": stats.get("rushing", {}).get("attempts"),
            "rush_yards": stats.get("rushing", {}).get("yards"),
            "rush_touchdowns": stats.get("rushing", {}).get("touchdowns"),
            "rush_avg": stats.get("rushing", {}).get("avg_yards"),
            "rush_first_downs": stats.get("rushing", {}).get("first_downs"),

            # -------- RECEIVING --------
            "rec_receptions": stats.get("receiving", {}).get("receptions"),
            "rec_yards": stats.get("receiving", {}).get("yards"),
            "rec_touchdowns": stats.get("receiving", {}).get("touchdowns"),
            "rec_avg": stats.get("receiving", {}).get("avg_yards"),

            # -------- KICK RETURNS --------
            "kr_returns": stats.get("kick_returns", {}).get("returns"),
            "kr_yards": stats.get("kick_returns", {}).get("yards"),
            "kr_avg": stats.get("kick_returns", {}).get("avg_yards"),

            # -------- FUMBLES --------
            "fumbles": stats.get("fumbles", {}).get("fumbles"),
            "lost_fumbles": stats.get("fumbles", {}).get("lost_fumbles"),
        }

        # -------- DEFENSE (only exists for 2024) --------
        defense = stats.get("defense", {})
        for k, v in defense.items():
            row[f"def_{k}"] = v  # create columns like def_tackles, def_sacks, etc.

        rows.append(row)

df_player_profile = pd.DataFrame(rows)
pd.set_option('display.max_columns', None)
df_player_profile

df_player_profile.info()

df_player_profile.to_csv("player_profile.csv", index=False, encoding="utf-8")
df_player_profile
```
