```python
rankingdata = data
rows = []

for team in rankingdata.get("rankings", []):
    rows.append({
        "poll_id": rankingdata["poll"].get("id"),
        "poll_name": rankingdata["poll"].get("name"),
        "poll_alias": rankingdata["poll"].get("alias"),
        "season": rankingdata.get("season"),
        "week": rankingdata.get("week"),
        "effective_time": rankingdata.get("effective_time"),

        "team_id": team.get("id"),
        "team_name": team.get("name"),
        "team_market": team.get("market"),
        "rank": team.get("rank"),
        "points": team.get("points"),
        "first_place_votes": team.get("fp_votes")
    })

df_rankings = pd.DataFrame(rows)
df_rankings 
```

```python
def explore_df(df, name="DataFrame"):
    print(f"\n==============================")
    print(f"üìä EXPLORING: {name}")
    print("==============================")

    # Shape
    print(f"\nüî∏ Shape: {df.shape[0]} rows √ó {df.shape[1]} columns")

    # Columns
    print("\nüî∏ Columns:")
    print(df.columns.tolist())

    # Data types
    print("\nüî∏ Data Types:")
    print(df.dtypes)

    # Missing values
    print("\nüî∏ Missing Values:")
    print(df.isnull().sum())

    # Memory usage
    print("\nüî∏ Memory Usage:")
    print(df.memory_usage(deep=True))

    # Head
    print("\nüî∏ First 5 Rows:")
    print(df.head())

    # Tail
    print("\nüî∏ Last 5 Rows:")
    print(df.tail())

    # Unique counts
    print("\nüî∏ Unique Values Per Column:")
    print(df.nunique())

    # Sample values per column
    print("\nüî∏ Sample Values Per Column:")
    for col in df.columns:
        print(f"  - {col}: {df[col].dropna().unique()[:5]}")

    print("\n==============================\n")
```

```python
explore_df(df_rankings, name="Rankings DataFrame")
```

```python
import pandas as pd
import glob
import os

# Columns you want
cols_to_extract = [
    "team_id",
    "name_id",
    "market",
    "alias",
    "founded",
    "mascot",
    "fight_song",
    "championships_won",
]

# Find all CSVs in root folder
files = glob.glob("*.csv")

# Output folder
output_folder = "data"
os.makedirs(output_folder, exist_ok=True)

combined_rows = []

print("\nüîç Checking CSV files...\n")

for f in files:
    df = pd.read_csv(f)

    # get only columns that actually exist in this file
    available_cols = [c for c in cols_to_extract if c in df.columns]

    if not available_cols:
        print(f"‚ö†Ô∏è Skipping {f} ‚Äî none of the target columns found")
        continue

    print(f"‚úî Extracting from: {f} ‚Üí {available_cols}")

    extracted = df[available_cols].copy()
    extracted["source_file"] = f

    combined_rows.append(extracted)

# Combine all extracted data
if combined_rows:
    final_df = pd.concat(combined_rows, ignore_index=True).drop_duplicates()

    save_path = os.path.join(output_folder, "combined_team_info.csv")
    final_df.to_csv(save_path, index=False)

    print(f"\nüéâ Saved extracted data to: {save_path}\n")
    print(final_df.head())
else:
    print("\n‚ùå No files contained any of the target columns.")
```
